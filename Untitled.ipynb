```python
# 1. Import Libraries
import pandas as pd
import numpy as np

# 2. Load the Data
df = pd.read_csv('Airbnb_Open_Data.csv')  # Update with the correct file path

# 3. Check for Missing Values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage[missing_percentage > 0].sort_values(ascending=False))

# 4. Handle Missing Values

# Fill missing categorical values with "Unknown"
categorical_cols = ["neighbourhood", "room type"]  # Update based on actual column names
for col in categorical_cols:
    df[col].fillna("Unknown", inplace=True)

# Clean 'price' column: Remove dollar sign and commas, then convert to numeric
df['price'] = df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Fill missing numerical values with median for numerical columns
numerical_cols = ["price", "number of reviews", "availability 365"]  # Update based on your dataset
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)

# 5. Remove Duplicates
df.drop_duplicates(inplace=True)

# 6. Handle Inconsistent Data: Convert categorical columns to lowercase (if needed)
df["room type"] = df["room type"].str.lower()  # Example: 'room type'
df["neighbourhood"] = df["neighbourhood"].str.lower()  # Example: 'neighbourhood'

# 7. Check for the final missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage[missing_percentage > 0].sort_values(ascending=False))

# 8. Check the cleaned data
print(df.head())

# 9. (Optional) Save the cleaned dataset to a new CSV file
df.to_csv('Cleaned_Airbnb_Open_Data.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting boxplots to detect outliers
numerical_cols = ["price", "number of reviews", "availability 365"]  # Columns to check for outliers
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Using IQR to detect outliers
Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Identifying outliers
outliers = ((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR)))
print(outliers.sum())  # This will print the number of outliers for each column
```

```python
# Cap the outliers
for col in numerical_cols:
    lower_limit = df[col].quantile(0.01)
    upper_limit = df[col].quantile(0.99)
    df[col] = np.clip(df[col], lower_limit, upper_limit)
```

```python
# Plot distribution of 'price'
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Boxplot for price by room type
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', data=df)
plt.title('Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Descriptive statistics for numerical columns
print(df.describe())
```

```python
# Group by 'room type' and calculate the average price
room_type_avg_price = df.groupby('room type')['price'].mean().sort_values(ascending=False)
print(room_type_avg_price)

# Plot the average price by room type with a different color palette
room_type_avg_price.plot(kind='bar', figsize=(10, 6), color=sns.color_palette('viridis', len(room_type_avg_price)))
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```
